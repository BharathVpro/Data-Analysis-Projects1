# Comparing Movies & Tv shows and their ratings

```python
plt.figure(figsize=(8,7))
df["type"].value_counts().plot(kind="pie",autopct="%1.1f%%",explode = [0,0.1])              #comparing show type
plt.title("PERCENTAGE OF MOVIES AND TV SHOWS")
plt.legend(loc = 'upper right', shadow = True)
plt.show()
```

```python
 ratings_ages = { 'TV-PG': 'Older Kids',                         #Converting Rating code into ages
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
```

```python
df["rating"]=df["rating"].replace(ratings_ages)                   #Replacing rating data with new data 
```

```python
plt.figure(figsize=(6.5,4.5))                                             
df['rating'].value_counts(normalize=True).plot.bar()             # ploting rating data using bar plot
plt.xticks(rotation=360)
plt.title('Distribution of rating categories')
plt.xlabel('rating')
plt.ylabel('relative frequency')
plt.show()
```

We noticed that Adults rating is on top & followed by Teen, older kids rating and Kids content is at the bottom

```python
plt.rcParams['figure.figsize'] = (8,6)
pd.crosstab(df["type"],
               df["rating"]).plot(kind="bar")                         #Spliting rating data by show type
plt.xticks(rotation=360)
plt.title("WHAT TYPE OF CONTENT IS MOST")
plt.show()
```

Highest number of rating is adult & kids is the lowest.

```python
early_ages =df[df['release_year'] > 2000]                            # Shrinking years data to early 20's
```

```python
plt.rcParams['figure.figsize'] = (12,7)
pd.crosstab(early_ages['release_year'],  
               early_ages["rating"]).sort_values(by='release_year',ascending=True).plot(kind="bar")
plt.title("RATINGS BASED ON EVERY YEAR")
plt.show()
```

There is a rise in 2018 year in all ratings. but, falling gradually from the year 2018.

```python
countries = df[(df['country']=='United States')|(df['country']=='India')|(df['country']=='United Kingdom')|(df['country']=='Japan')|
                             (df['country']=='Canada')|(df['country']=='Spain')]
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='country',                                                  #Country wise Netflix content based on show type
                 hue='type',
                 data=countries,
                 palette="magma")
plt.title('comparing between the show types that the top countries produce')
plt.show()
```

US and India are top in produces movies and Tv shows in that Movies are high.

Japan, Canada, Spain are produces low movies & Tv shows But, here they more TV shows than movies.


```python
plt.rcParams['figure.figsize'] = (10,7)
pd.crosstab(df["type"],                                                           #Spliting rating data by show type
               df["rating"]).plot(kind="bar")
plt.xticks(rotation=360)
plt.title("WHAT TYPE OF CONTENT IS MOST AVAILABLE")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = df["listed_in"].value_counts().head(10).index,                     #excuting Top Genre from dataset
              y = df["listed_in"].value_counts().head(10).values,
            palette="rocket")
plt.xticks(rotation=50)
plt.title("Top10 Genre watched from all the countries",fontweight="bold")
plt.show()
```


# Netflix in India

```python
netflix_india = df[df.country == 'India']                                    #Extracting Indian content from dataset

early_india = netflix_india[netflix_india['release_year']>2000]          #Shrinking years to early 20's
```

```python
plt.rcParams['figure.figsize'] = (15,8)
pd.crosstab(early_india["release_year"],
               early_india["rating"]).sort_values(by="release_year",ascending=False).plot(kind="bar")
plt.xticks(rotation=360)
plt.title("RATINGS BASED ON YEAR FOR NETFLIX INDIA")
plt.show()
```

There is a rise of netflix content from 2001 to 2017. but, falling from the year 2017.

```python
plt.rcParams['figure.figsize'] = (8,6)
pd.crosstab(netflix_india["type"],
            netflix_india["rating"]).plot(kind="bar")                              #Plottiing netflix_india show type content
plt.xticks(rotation=360)
plt.title("WHAT TYPE OF CONTENT IS MOST WATCHED")
plt.show()
```

It shows that Netflix india produces Movies more than Tv show with high Teen content & low kids content.

```python
plt.figure(figsize=(15,5))                                                                     # Plotting netflix_india genres
sns.barplot(x = netflix_india["listed_in"].value_counts().head(10).index,
              y = netflix_india["listed_in"].value_counts().head(10).values,palette="husl")
plt.xticks(rotation=50)
plt.title("Top10 Genre watched by Indian Users",fontweight="bold")
plt.show()
```
